```python
INPUT_DIR = "data/layouts"
OUTPUT_FILE = "data/layouts/import_label_studio.json"
OUTPUT_DATASET_FILE = "data/layouts/import_label_studio.json"

def convert_to_ls(image_path):
    img = Image.open(image_path)
    w, h = img.size

    # Récupérer les données détaillées de Tesseract (dict)
    data = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)

    results = []

    for i in range(len(data['text'])):
        # On ne garde que les blocs avec du texte et une confiance > 0
        if int(data['conf'][i]) > 0 and data['text'][i].strip():
            # Conversion des pixels en pourcentages Label Studio
            x = (data['left'][i] / w) * 100
            y = (data['top'][i] / h) * 100
            width = (data['width'][i] / w) * 100
            height = (data['height'][i] / h) * 100

            # ID unique pour lier la BBox et le Texte
            region_id = f"region_{i}"

            # 1. Création de la Bounding Box
            results.append({
                "original_width": w,
                "original_height": h,
                "image_rotation": 0,
                "value": {
                    "x": x, "y": y,
                    "width": width, "height": height,
                    "rotation": 0,
                    "rectanglelabels": ["Texte"]
                },
                "id": region_id,
                "from_name": "bbox",
                "to_name": "image",
                "type": "rectanglelabels"
            })

            # 2. Création de la Transcription (Text Area)
            results.append({
                "original_width": w,
                "original_height": h,
                "value": {
                    "x": x, "y": y,
                    "width": width, "height": height,
                    "rotation": 0,
                    "text": [data['text'][i]]
                },
                "id": region_id,
                "from_name": "transcription",
                "to_name": "image",
                "type": "textarea"
            })

    return {
        "data": {
            "ocr": f"/data/local-files/?d={image_path.replace(os.sep, '/')}"
        },
        "predictions": [{"model_version": "tesseract", "result": results}]
    }

# Main loop
all_tasks = []
for img_file in glob.glob(f"{INPUT_DIR}/*.jpg") + glob.glob(f"{INPUT_DIR}/*.png"):
    print(f"Traitement de : {img_file}")
    all_tasks.append(convert_to_ls(img_file))

with open(OUTPUT_FILE, 'w', encoding='utf-8') as f:
    json.dump(all_tasks, f, indent=2, ensure_ascii=False)

print(f"\nFichier prêt ! Importez '{OUTPUT_FILE}' dans Label Studio.")
```

```python
def normalize_bbox(bbox, width, height):
    """Normalise les coordonnées (x, y, w, h) sur une échelle de 0-1000."""
    return [
        int(1000 * (bbox[0] / width)),        # x_min
        int(1000 * (bbox[1] / height)),       # y_min
        int(1000 * ((bbox[0] + bbox[2]) / width)), # x_max
        int(1000 * ((bbox[1] + bbox[3]) / height))  # y_max
    ]
```

```python
def get_layoutlm_features(image_path,label_id):
    img = Image.open(image_path).convert("RGB")
    width, height = img.size

    # Extraction des données avec Tesseract
    ocr_data = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)

    words = []
    bboxes = []

    for i in range(len(ocr_data['text'])):
        conf = int(ocr_data['conf'][i])
        text = ocr_data['text'][i].strip()

        if conf > 0 and text:
            # Coordonnées brutes
            l, t, w, h = ocr_data['left'][i], ocr_data['top'][i], ocr_data['width'][i], ocr_data['height'][i]

            # Normalisation pour LayoutLM (0 à 1000)
            # Format attendu : [x_min, y_min, x_max, y_max]
            x0 = int(1000 * (l / width))
            y0 = int(1000 * (t / height))
            x1 = int(1000 * ((l + w) / width))
            y1 = int(1000 * ((t + h) / height))

            # Sécurité pour ne pas dépasser 1000
            bboxes.append([
                min(max(0, x0), 1000),
                min(max(0, y0), 1000),
                min(max(0, x1), 1000),
                min(max(0, y1), 1000)
            ])
            words.append(text)

    return {"words": words, "bboxes": bboxes, "image_path": image_path, "label": label_id }

# Traitement par lots
dataset = []
i = 1
for document in documents:
    print(f"Normalisation de { document["path"]} pour LayoutLM...")
    dataset.append(get_layoutlm_features(document["path"],i))
    i= i + 1

# Sauvegarde pour l'entraînement
with open("layoutlm_input.json", "w", encoding="utf-8") as f:
    json.dump(dataset, f, indent=2, ensure_ascii=False)
```

```python
import torch
import json
import os
from torch.optim import AdamW
from PIL import Image
from transformers import (
    LayoutLMv3ForSequenceClassification,
    LayoutLMv3Processor,
    LayoutLMv3ImageProcessor
)

# ==========================================
# 1. ENCODEUR PERSONNALISÉ (VOTRE SOLUTION)
# ==========================================
class LayoutLMv3Encoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, LayoutLMv3ImageProcessor):
            return obj.to_dict()  # Utilise la conversion intégrée de Hugging Face
        return super().default(obj)

# 2. CONFIGURATION
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_id = "microsoft/layoutlmv3-base"
output_dir = "./layoutlmv3_finetuned"

# Initialisation du processeur
image_processor = LayoutLMv3ImageProcessor(apply_ocr=False)
processor = LayoutLMv3Processor.from_pretrained(model_id, image_processor=image_processor)

# Modèle pour 6 classes
model = LayoutLMv3ForSequenceClassification.from_pretrained(model_id, num_labels=6)
model.to(device)

# 3. PRÉPARATION DES DONNÉES
with open("layoutlm_input.json", "r", encoding="utf-8") as f:
    data = json.load(f)

def prepare_inputs(item, label):
    # LayoutLMv3 nécessite une image pour l'encodage visuel (patchs)
    dummy_img = Image.new("RGB", (224, 224), color="white")

    encoding = processor(
        dummy_img,
        text=item["words"],
        boxes=item["bboxes"],
        return_tensors="pt",
        truncation=True,
        padding="max_length",
        max_length=512
    )

    return {
        "input_ids": encoding["input_ids"].to(device),
        "attention_mask": encoding["attention_mask"].to(device),
        "bbox": encoding["bbox"].to(device),
        "pixel_values": encoding["pixel_values"].to(device),
        "labels": torch.tensor([label]).to(device)
    }

# 4. BOUCLE D'ENTRAÎNEMENT
optimizer = AdamW(model.parameters(), lr=2e-5)
model.train()

print(f"🚀 Entraînement LayoutLMv3 démarré sur {device}...")

for i, item in enumerate(data):
    try:
        optimizer.zero_grad()
        label = int(item.get("label", 0))
        inputs = prepare_inputs(item, label)

        outputs = model(**inputs)
        loss = outputs.loss
        loss.backward()
        optimizer.step()

        if i % 10 == 0:
            print(f"Batch {i}/{len(data)} | Loss: {loss.item():.4f}")
    except Exception as e:
        print(f"⚠️ Erreur doc {i}: {e}")

import pickle
import os

# 5. SAUVEGARDE VIA PICKLE (SOLUTION ULTIME)
print("💾 Sauvegarde hybride (Modèle standard + Processeur Pickle)...")
os.makedirs(output_dir, exist_ok=True)

# A. Sauvegarde du modèle (Format standard Hugging Face)
model.save_pretrained(output_dir)

# B. Sauvegarde du processeur complet avec Pickle
# On sauvegarde TOUT l'objet 'processor' d'un coup
processor_path = os.path.join(output_dir, "full_processor.pkl")
with open(processor_path, "wb") as f:
    pickle.dump(processor, f)
```

```python
import torch
import pytesseract
from PIL import Image
from transformers import LayoutLMv3Processor, LayoutLMv3ForTokenClassification

# 1. Chargement (Adaptez le chemin vers votre dossier de sauvegarde)
model_path = "./layoutlmv3_finetuned"
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# On charge le processeur (via le fichier processor_config.json que nous avons créé)
# Rechargement du processeur via Pickle
with open(os.path.join(output_dir, "full_processor.pkl"), "rb") as f:
    processor = pickle.load(f)
model = LayoutLMv3ForTokenClassification.from_pretrained(model_path)
model.to(device)
model.eval()

def extract_entities(image_path):
    img = Image.open(image_path).convert("RGB")
    width, height = img.size

    # 2. OCR via Tesseract
    ocr_data = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)

    words = []
    bboxes = []
    for i in range(len(ocr_data['text'])):
        word = ocr_data['text'][i]
        if word.strip(): # Nettoyage des espaces vides
            words.append(word)
            # Normalisation 0-1000 pour LayoutLM
            x, y, w, h = ocr_data['left'][i], ocr_data['top'][i], ocr_data['width'][i], ocr_data['height'][i]
            bbox = [
                int(1000 * (x / width)),
                int(1000 * (y / height)),
                int(1000 * ((x + w) / width)),
                int(1000 * ((y + h) / height))
            ]
            bboxes.append(bbox)

    # 3. Encodage (LayoutLMv3 a besoin de l'image pour pixel_values)
    encoding = processor(img, words, boxes=bboxes, return_tensors="pt", truncation=True, padding="max_length")
    encoding = {k: v.to(device) for k, v in encoding.items()}

    # 4. Prédiction
    with torch.no_grad():
        outputs = model(**encoding)

    # Récupération des probabilités les plus hautes
    logits = outputs.logits
    predictions = logits.argmax(-1).squeeze().tolist()

    # 5. Mapping et Alignement
    # Important : On utilise word_ids pour savoir quel token appartient à quel mot
    word_ids = encoding.word_ids(batch_index=0)
    id2label = model.config.id2label # Récupère les labels sauvegardés dans le modèle

    results = []
    last_word_id = None

    for idx, word_id in enumerate(word_ids):
        if word_id is None or word_id == last_word_id:
            continue # On ignore les tokens spéciaux ([CLS], [SEP]) et les sub-tokens (##ure)

        label_id = predictions[idx]
        if label_id != 0: # Si différent de "O"
            label_name = id2label[label_id]
            results.append({
                "word": words[word_id],
                "label": label_name,
                "bbox": bboxes[word_id]
            })
        last_word_id = word_id

    return results

# Test
entities = extract_entities("votre_facture.jpg")
for ent in entities:
    print(f"[{ent['label']}] : {ent['word']} (Coordonnées: {ent['bbox']})")
```

```python
import os
import json
import torch
import pytesseract
from pathlib import Path
from PIL import Image
from pdf2image import convert_from_path
from tqdm import tqdm

# --- CONFIGURATION ---
INPUT_DIR = Path("data/layouts")
OUTPUT_IMAGE_DIR = INPUT_DIR / "processed_images"
OUTPUT_JSON = "layoutlm_input.json"
POPPLER_PATH = None  # Spécifiez le chemin si nécessaire (ex: r"C:\poppler\bin")
DPI = 200            # Équilibre parfait entre vitesse et précision OCR

# Création du dossier de sortie pour les images
OUTPUT_IMAGE_DIR.mkdir(parents=True, exist_ok=True)

def normalize_bbox(bbox, width, height):
    """Normalise [x, y, w, h] vers [x0, y0, x1, y1] sur l'échelle 0-999."""
    x0 = max(0, min(int(1000 * (bbox[0] / width)), 999))
    y0 = max(0, min(int(1000 * (bbox[1] / height)), 999))
    x1 = max(0, min(int(1000 * ((bbox[0] + bbox[2]) / width)), 999))
    y1 = max(0, min(int(1000 * ((bbox[1] + bbox[3]) / height)), 999))
    return [x0, y0, x1, y1]

def process_pdf_to_layoutlm(pdf_path, label_id):
    """Convertit un PDF en images, fait l'OCR et extrait les features."""
    results = []
    try:
        # 1. Conversion PDF en Images
        pages = convert_from_path(pdf_path, dpi=DPI, poppler_path=POPPLER_PATH)

        for i, page in enumerate(pages):
            image_name = f"{pdf_path.stem}_p{i}.jpg"
            image_path = OUTPUT_IMAGE_DIR / image_name
            page.save(image_path, "JPEG")

            # 2. OCR avec Tesseract
            width, height = page.size
            ocr_data = pytesseract.image_to_data(page, output_type=pytesseract.Output.DICT)

            words, bboxes = [], []
            for j in range(len(ocr_data['text'])):
                text = ocr_data['text'][j].strip()
                conf = int(ocr_data['conf'][j])

                if conf > 15 and text: # Seuil de confiance pour filtrer le bruit
                    raw_box = [
                        ocr_data['left'][j], ocr_data['top'][j],
                        ocr_data['width'][j], ocr_data['height'][j]
                    ]
                    norm_box = normalize_bbox(raw_box, width, height)

                    if norm_box[2] > norm_box[0] and norm_box[3] > norm_box[1]:
                        words.append(text)
                        bboxes.append(norm_box)

            if words:
                results.append({
                    "id": f"{pdf_path.stem}_{i}",
                    "words": words,
                    "bboxes": bboxes,
                    "image_path": str(image_path),
                    "label": label_id
                })
    except Exception as e:
        print(f"\n❌ Erreur sur {pdf_path.name}: {e}")

    return results

# --- BOUCLE PRINCIPALE ---
all_data = []
pdf_files = list(INPUT_DIR.rglob("*.pdf"))

print(f"🚀 Début du traitement de {len(pdf_files)} fichiers...")

for pdf in tqdm(pdf_files):
    # Ici, vous pouvez définir une logique pour le label_id
    # (ex: 1 si 'facture' est dans le nom, sinon 0)
    label = 1 if "facture" in pdf.name.lower() else 0

    page_features = process_pdf_to_layoutlm(pdf, label)
    all_data.extend(page_features)

# Sauvegarde finale
with open(OUTPUT_JSON, "w", encoding="utf-8") as f:
    json.dump(all_data, f, indent=2, ensure_ascii=False)

print(f"\n✅ Terminé ! {len(all_data)} pages traitées et sauvegardées dans {OUTPUT_JSON}")
```

```python
import os
import json
import torch
import pytesseract
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image
from pdf2image import convert_from_path
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset, DataLoader
from transformers import LayoutLMForSequenceClassification, LayoutLMTokenizer
from torch.optim import AdamW
# --- 1. CONFIGURATION ---
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MODEL_NAME = "microsoft/layoutlm-base-uncased"
INPUT_DIR = Path("data/layouts")
OUTPUT_IMAGE_DIR = INPUT_DIR / "processed_images"
OUTPUT_IMAGE_DIR.mkdir(parents=True, exist_ok=True)

EPOCHS = 5
BATCH_SIZE = 4
LR = 2e-5

# --- 2. FONCTIONS DE PRÉPARATION (OCR & NORMALISATION) ---
def normalize_bbox(bbox, width, height):
    """Normalise [x, y, w, h] vers [x0, y0, x1, y1] sur l'échelle 0-999."""
    x0 = max(0, min(int(1000 * (bbox[0] / width)), 999))
    y0 = max(0, min(int(1000 * (bbox[1] / height)), 999))
    x1 = max(0, min(int(1000 * ((bbox[0] + bbox[2]) / width)), 999))
    y1 = max(0, min(int(1000 * ((bbox[1] + bbox[3]) / height)), 999))
    return [x0, y0, x1, y1]

def extract_features(pdf_path):
    """Convertit PDF -> Image -> OCR -> Features LayoutLM."""
    results = []
    try:
        # Conversion DPI 200 pour un bon compromis vitesse/précision
        pages = convert_from_path(pdf_path, dpi=200)
        # Label 1 si 'facture' est dans le nom du fichier, sinon 0
        label = 1 if "facture" in pdf_path.name.lower() else 0

        for i, page in enumerate(pages):
            img_path = OUTPUT_IMAGE_DIR / f"{pdf_path.stem}_p{i}.jpg"
            page.save(img_path, "JPEG")

            w, h = page.size
            ocr = pytesseract.image_to_data(page, output_type=pytesseract.Output.DICT)

            words, bboxes = [], []
            for j in range(len(ocr['text'])):
                text = ocr['text'][j].strip()
                if int(ocr['conf'][j]) > 15 and text:
                    box = normalize_bbox([ocr['left'][j], ocr['top'][j], ocr['width'][j], ocr['height'][j]], w, h)
                    words.append(text)
                    bboxes.append(box)

            if words:
                results.append({"words": words, "bboxes": bboxes, "label": label})
    except Exception as e:
        print(f"Erreur sur {pdf_path.name}: {e}")
    return results

# --- 3. CLASSE DATASET ROBUSTE ---
class LayoutDataset(Dataset):
    def __init__(self, data, tokenizer):
        self.data = data
        self.tokenizer = tokenizer

    def __len__(self): return len(self.data)

    def __getitem__(self, idx):
        item = self.data[idx]
        # Le tokenizer gère l'alignement auto des bboxes avec les sous-tokens
        encoding = self.tokenizer(
            item["words"],
            boxes=item["bboxes"],
            truncation=True,
            padding="max_length",
            max_length=512,
            return_tensors="pt"
        )

        return {
            "input_ids": encoding["input_ids"].squeeze(0).to(DEVICE),
            "bbox": torch.tensor([item["bboxes"]]) ,
            "attention_mask": encoding["attention_mask"].squeeze(0).to(DEVICE),
            "token_type_ids": encoding["token_type_ids"].squeeze(0).to(DEVICE),
            "labels": torch.tensor(item["label"]).to(DEVICE)
        }

# --- 4. PIPELINE D'ENTRAÎNEMENT ---
print("🔍 Étape 1 : Extraction OCR et préparation des données...")
all_docs = []
pdf_files = list(INPUT_DIR.rglob("*.pdf"))
for pdf in tqdm(pdf_files):
    all_docs.extend(extract_features(pdf))

# Séparation Train (80%) / Val (20%)
train_data, val_data = train_test_split(all_docs, test_size=0.2, random_state=42)

tokenizer = LayoutLMTokenizer.from_pretrained(MODEL_NAME)
train_loader = DataLoader(LayoutDataset(train_data, tokenizer), batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(LayoutDataset(val_data, tokenizer), batch_size=BATCH_SIZE)

# Chargement du modèle
model = LayoutLMForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=2).to(DEVICE)
optimizer = AdamW(model.parameters(), lr=LR)

history = {"train_loss": [], "val_acc": []}

print(f"🚀 Étape 2 : Entraînement sur {DEVICE}...")
for epoch in range(EPOCHS):
    model.train()
    total_loss = 0
    for batch in tqdm(train_loader, leave=False):
        optimizer.zero_grad()
        outputs = model(**batch)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    # Évaluation
    model.eval()
    correct = 0
    with torch.no_grad():
        for batch in val_loader:
            outputs = model(**batch)
            preds = outputs.logits.argmax(-1)
            correct += (preds == batch["labels"]).sum().item()

    acc = correct / len(val_data)
    history["train_loss"].append(total_loss / len(train_loader))
    history["val_acc"].append(acc)

    print(f"Epoch {epoch+1}/{EPOCHS} | Loss: {history['train_loss'][-1]:.4f} | Acc Val: {acc:.2%}")

# --- 5. SAUVEGARDE ET VISUALISATION ---
model.save_pretrained("./layoutlm_final_model")
tokenizer.save_pretrained("./layoutlm_final_model")



plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history["train_loss"], label="Train Loss")
plt.title("Perte d'entraînement")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history["val_acc"], label="Validation Acc", color="green")
plt.title("Précision (Accuracy)")
plt.legend()
plt.show()

print("✅ Pipeline terminé. Modèle prêt dans './layoutlm_final_model'")
```

```python
import os
import torch
import pytesseract
import matplotlib.pyplot as plt
from pathlib import Path
from PIL import Image
from pdf2image import convert_from_path
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset, DataLoader
from transformers import LayoutLMForSequenceClassification, LayoutLMTokenizer
from torch.optim import AdamW

# --- 1. CONFIGURATION ---
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MODEL_NAME = "microsoft/layoutlm-base-uncased"
INPUT_DIR = Path("data/layouts")
OUTPUT_IMAGE_DIR = INPUT_DIR / "processed_images"
OUTPUT_IMAGE_DIR.mkdir(parents=True, exist_ok=True)

MAX_SEQ_LEN = 512
EPOCHS = 5
BATCH_SIZE = 4
LR = 2e-5

# --- 2. LOGIQUE MANUELLE D'ALIGNEMENT ET PADDING ---
def prepare_manual_tokens(tokenizer, words, bboxes, max_len=512):
    """
    Gère manuellement l'alignement Sub-word :
    Si 'Facturation' devient ['Fact', '##uration'], la bbox est dupliquée.
    """
    token_ids = []
    token_boxes = []

    for word, box in zip(words, bboxes):
        # On récupère les IDs des sous-tokens sans les tokens spéciaux [CLS]/[SEP]
        word_tokens = tokenizer.encode(word, add_special_tokens=False)

        # Extension manuelle : on duplique la box pour chaque morceau de mot
        token_ids.extend(word_tokens)
        token_boxes.extend([box] * len(word_tokens))

    # Troncature manuelle (on garde de la place pour CLS et SEP)
    if len(token_ids) > max_len - 2:
        token_ids = token_ids[:max_len - 2]
        token_boxes = token_boxes[:max_len - 2]

    # Ajout manuel des tokens spéciaux : CLS (début) et SEP (fin)
    # On utilise [0,0,0,0] pour les positions non-visuelles
    token_ids = [tokenizer.cls_token_id] + token_ids + [tokenizer.sep_token_id]
    token_boxes = [[0, 0, 0, 0]] + token_boxes + [[0, 0, 0, 0]]

    # Masque d'attention manuel (1 pour les vrais tokens, 0 pour le padding)
    attention_mask = [1] * len(token_ids)

    # PADDING manuel jusqu'à 512 pour que le DataLoader puisse "stacker" les batchs
    pad_len = max_len - len(token_ids)
    token_ids += [tokenizer.pad_token_id] * pad_len
    token_boxes += [[0, 0, 0, 0]] * pad_len
    attention_mask += [0] * pad_len

    return {
        "input_ids": torch.tensor(token_ids, dtype=torch.long),
        "bbox": torch.tensor(token_boxes, dtype=torch.long),
        "attention_mask": torch.tensor(attention_mask, dtype=torch.long)
    }

# --- 3. EXTRACTION ET OCR ---
def normalize_bbox(bbox, w, h):
    """Normalisation stricte 0-999 pour éviter les IndexError."""
    return [
        max(0, min(int(1000 * (bbox[0] / w)), 999)),
        max(0, min(int(1000 * (bbox[1] / h)), 999)),
        max(0, min(int(1000 * ((bbox[0] + bbox[2]) / w)), 999)),
        max(0, min(int(1000 * ((bbox[1] + bbox[3]) / h)), 999))
    ]

def process_pdfs():
    data = []
    for pdf in list(INPUT_DIR.rglob("*.pdf")):
        try:
            pages = convert_from_path(pdf, dpi=200)
            label = 1 if "facture" in pdf.name.lower() else 0
            for i, page in enumerate(pages):
                w, h = page.size
                ocr = pytesseract.image_to_data(page, output_type=pytesseract.Output.DICT)
                words, bboxes = [], []
                for j, text in enumerate(ocr['text']):
                    if text.strip() and int(ocr['conf'][j]) > 15:
                        words.append(text)
                        bboxes.append(normalize_bbox([ocr['left'][j], ocr['top'][j], ocr['width'][j], ocr['height'][j]], w, h))
                if words: data.append({"words": words, "bboxes": bboxes, "label": label})
        except Exception as e: print(f"Erreur {pdf}: {e}")
    return data

# --- 4. DATASET & ENTRAÎNEMENT ---
class LayoutDataset(Dataset):
    def __init__(self, data, tokenizer):
        self.data, self.tokenizer = data, tokenizer
    def __len__(self): return len(self.data)
    def __getitem__(self, idx):
        item = self.data[idx]
        feat = prepare_manual_tokens(self.tokenizer, item["words"], item["bboxes"])
        feat["labels"] = torch.tensor(item["label"], dtype=torch.long)
        return {k: v.to(DEVICE) for k, v in feat.items()}

# Initialisation
all_data = process_pdfs()
train_data, val_data = train_test_split(all_docs, test_size=0.2)
tokenizer = LayoutLMTokenizer.from_pretrained(MODEL_NAME)
train_loader = DataLoader(LayoutDataset(train_data, tokenizer), batch_size=BATCH_SIZE, shuffle=True)
val_loader = DataLoader(LayoutDataset(val_data, tokenizer), batch_size=BATCH_SIZE)

model = LayoutLMForSequenceClassification.from_pretrained(MODEL_NAME, num_labels=2).to(DEVICE)
optimizer = AdamW(model.parameters(), lr=LR)

# Boucle d'entraînement
for epoch in range(EPOCHS):
    model.train()
    for batch in tqdm(train_loader):
        optimizer.zero_grad()
        loss = model(**batch).loss
        loss.backward()
        optimizer.step()

    # Validation
    model.eval()
    correct = 0
    with torch.no_grad():
        for batch in val_loader:
            outputs = model(**batch)
            correct += (outputs.logits.argmax(-1) == batch["labels"]).sum().item()
    print(f"Epoch {epoch+1} | Acc Val: {correct/len(val_data):.2%}")

model.save_pretrained("./layoutlm_manual_final")
```

```python
import torch
import pytesseract
from PIL import Image, ImageDraw
from transformers import LayoutLMForSequenceClassification, LayoutLMTokenizer

# --- CONFIGURATION ---
MODEL_PATH = "./layoutlm_manual_final"
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Chargement
tokenizer = LayoutLMTokenizer.from_pretrained(MODEL_PATH)
model = LayoutLMForSequenceClassification.from_pretrained(MODEL_PATH).to(DEVICE)
model.eval()

def run_inference_and_visualize(image_path):
    # 1. OCR et Chargement
    image = Image.open(image_path).convert("RGB")
    width, height = image.size
    ocr = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)

    words, bboxes = [], []
    draw = ImageDraw.Draw(image)

    # 2. Extraction & Normalisation (Stricte 0-999)
    for i, text in enumerate(ocr['text']):
        if text.strip() and int(ocr['conf'][i]) > 15:
            l, t, w, h = ocr['left'][i], ocr['top'][i], ocr['width'][i], ocr['height'][i]

            norm_box = [
                max(0, min(int(1000 * (l / width)), 999)),
                max(0, min(int(1000 * (t / height)), 999)),
                max(0, min(int(1000 * ((l + w) / width)), 999)),
                max(0, min(int(1000 * ((t + h) / height)), 999))
            ]
            words.append(text)
            bboxes.append(norm_box)

            # Dessin des boîtes OCR en rouge pour vérification
            draw.rectangle([l, t, l + w, t + h], outline="red", width=2)

    # 3. Préparation Manuelle (Alignement Sub-tokens + Padding 512)
    # On réutilise la même logique que pour l'entraînement
    inputs = prepare_manual_tokens(tokenizer, words, bboxes, max_len=512)

    # Ajout de la dimension Batch (1, 512, ...)
    input_ids = inputs["input_ids"].unsqueeze(0).to(DEVICE)
    bbox = inputs["bbox"].unsqueeze(0).to(DEVICE)
    attention_mask = inputs["attention_mask"].unsqueeze(0).to(DEVICE)

    # 4. Prédiction
    with torch.no_grad():
        outputs = model(input_ids=input_ids, bbox=bbox, attention_mask=attention_mask)
        logits = outputs.logits
        idx = torch.argmax(logits, dim=-1).item()

    label_name = "FACTURE" if idx == 1 else "AUTRE"
    confidence = torch.softmax(logits, dim=-1)[0][idx].item()

    # 5. Affichage
    print(f"🔍 Résultat : {idx}  {label_name} ({confidence:.2%})")

    # Annoter l'image
    draw.text((10, 10), f"CLASSE: {label_name} | CONF: {confidence:.2%}", fill="green")
    image.save("resultat_inference.jpg")
    image.show()

# Utilisation :
run_inference_and_visualize("data/layouts/test_1.jpg")
```

```python
import csv
import torch
from pathlib import Path
from tqdm import tqdm
from pdf2image import convert_from_path
import pytesseract
from transformers import LayoutLMForSequenceClassification, LayoutLMTokenizer

# --- CONFIGURATION ---
MODEL_PATH = "./layoutlm_manual_final"
INPUT_FOLDER = Path("data/to_predict")  # Dossier contenant les nouveaux PDF
OUTPUT_CSV = "predictions_rapport.csv"
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Chargement du modèle et du tokenizer
tokenizer = LayoutLMTokenizer.from_pretrained(MODEL_PATH)
model = LayoutLMForSequenceClassification.from_pretrained(MODEL_PATH).to(DEVICE)
model.eval()

def process_and_predict(pdf_path):
    """Prédit la classe pour chaque page d'un PDF."""
    results = []
    try:
        pages = convert_from_path(pdf_path, dpi=200)
        for i, page in enumerate(pages):
            w, h = page.size
            ocr = pytesseract.image_to_data(page, output_type=pytesseract.Output.DICT)

            words, bboxes = [], []
            for j, text in enumerate(ocr['text']):
                if text.strip() and int(ocr['conf'][j]) > 15:
                    l, t, width, height = ocr['left'][j], ocr['top'][j], ocr['width'][j], ocr['height'][j]
                    norm_box = [
                        max(0, min(int(1000 * (l / w)), 999)),
                        max(0, min(int(1000 * (t / h)), 999)),
                        max(0, min(int(1000 * ((l + width) / w)), 999)),
                        max(0, min(int(1000 * ((t + height) / h)), 999))
                    ]
                    words.append(text)
                    bboxes.append(norm_box)

            if not words: continue

            # Utilisation de votre fonction manuelle (doit être définie dans le script)
            inputs = prepare_manual_tokens(tokenizer, words, bboxes)

            with torch.no_grad():
                outputs = model(
                    input_ids=inputs["input_ids"].unsqueeze(0).to(DEVICE),
                    bbox=inputs["bbox"].unsqueeze(0).to(DEVICE),
                    attention_mask=inputs["attention_mask"].unsqueeze(0).to(DEVICE)
                )

            probs = torch.softmax(outputs.logits, dim=-1)
            pred_idx = torch.argmax(probs, dim=-1).item()
            conf = probs[0][pred_idx].item()

            results.append({
                "filename": pdf_path.name,
                "page": i + 1,
                "prediction": "FACTURE" if pred_idx == 1 else "AUTRE",
                "confidence": round(conf, 4)
            })
    except Exception as e:
        print(f"Erreur sur {pdf_path}: {e}")
    return results

# --- BOUCLE PRINCIPALE ---
all_predictions = []
files = list(INPUT_FOLDER.glob("*.pdf"))

print(f"🔎 Analyse de {len(files)} documents...")
for pdf in tqdm(files):
    all_predictions.extend(process_and_predict(pdf))

# --- EXPORTATION CSV ---
keys = ["filename", "page", "prediction", "confidence"]
with open(OUTPUT_CSV, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=keys)
    writer.writeheader()
    writer.writerows(all_predictions)

print(f"✅ Rapport généré : {OUTPUT_CSV}")
```

```python
import os
import csv
import torch
import pytesseract
from PIL import Image
from pathlib import Path
from tqdm import tqdm
from torch.utils.data import Dataset, DataLoader
from transformers import LayoutLMForSequenceClassification, LayoutLMTokenizer
from torch.optim import AdamW

# --- 1. CONFIGURATION ---
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MODEL_NAME = "microsoft/layoutlm-base-uncased"
MAX_SEQ_LEN = 512

# --- 2. LOGIQUE MANUELLE D'ALIGNEMENT (ESSENTIEL) ---
def prepare_manual_tokens(tokenizer, words, bboxes, max_len=512):
    """Alignement manuel des sous-tokens et padding strict à 512."""
    token_ids, token_boxes = [], []

    for word, box in zip(words, bboxes):
        word_tokens = tokenizer.encode(word, add_special_tokens=False)
        token_ids.extend(word_tokens)
        token_boxes.extend([box] * len(word_tokens)) # Duplication de la box pour chaque token

    # Troncature
    token_ids = [tokenizer.cls_token_id] + token_ids[:max_len-2] + [tokenizer.sep_token_id]
    token_boxes = [[0, 0, 0, 0]] + token_boxes[:max_len-2] + [[0, 0, 0, 0]]

    # Masque d'attention et Padding
    pad_len = max_len - len(token_ids)
    attention_mask = [1] * len(token_ids) + [0] * pad_len
    token_ids += [tokenizer.pad_token_id] * pad_len
    token_boxes += [[0, 0, 0, 0]] * pad_len

    return {
        "input_ids": torch.tensor(token_ids, dtype=torch.long),
        "bbox": torch.tensor(token_boxes, dtype=torch.long),
        "attention_mask": torch.tensor(attention_mask, dtype=torch.long)
    }

# --- 3. DATASET ADAPTÉ À VOTRE LISTE ---
class CustomLayoutDataset(Dataset):
    def __init__(self, data_list, tokenizer):
        self.data_list = data_list
        self.tokenizer = tokenizer

    def __len__(self): return len(self.data_list)

    def __getitem__(self, idx):
        item = self.data_list[idx]
        image = Image.open(item['path']).convert("RGB")
        w, h = image.size

        # OCR
        ocr = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)
        words, bboxes = [], []
        for j, text in enumerate(ocr['text']):
            if text.strip() and int(ocr['conf'][j]) > 15:
                # Normalisation 0-999
                box = [
                    max(0, min(int(1000 * (ocr['left'][j] / w)), 999)),
                    max(0, min(int(1000 * (ocr['top'][j] / h)), 999)),
                    max(0, min(int(1000 * ((ocr['left'][j] + ocr['width'][j]) / w)), 999)),
                    max(0, min(int(1000 * ((ocr['top'][j] + ocr['height'][j]) / h)), 999))
                ]
                words.append(text)
                bboxes.append(box)

        features = prepare_manual_tokens(self.tokenizer, words, bboxes)

        # On définit le label dynamiquement pour l'entraînement (exemple ici)
        label = 1 if "Drug_Influence" in item['path'] else 0
        features["labels"] = torch.tensor(label, dtype=torch.long)

        return {k: v.to(DEVICE) for k, v in features.items()}

# --- 4. EXPORTATION DES PRÉDICTIONS (RAPPORT FINAL) ---
def generate_report(doc_list, model, tokenizer, output_file="rapport_final.csv"):
    model.eval()
    results = []

    print(f"📊 Génération du rapport pour {len(doc_list)} images...")

    for item in tqdm(doc_list):
        # Utilisation de l'image via son path
        image = Image.open(item['path']).convert("RGB")
        w, h = image.size
        ocr = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)

        words, bboxes = [], []
        for j, text in enumerate(ocr['text']):
            if text.strip() and int(ocr['conf'][j]) > 15:
                bboxes.append([
                    max(0, min(int(1000 * (ocr['left'][j] / w)), 999)),
                    max(0, min(int(1000 * (ocr['top'][j] / h)), 999)),
                    max(0, min(int(1000 * ((ocr['left'][j] + ocr['width'][j]) / w)), 999)),
                    max(0, min(int(1000 * ((ocr['top'][j] + ocr['height'][j]) / h)), 999))
                ])
                words.append(text)

        if not words: continue

        feat = prepare_manual_tokens(tokenizer, words, bboxes)

        with torch.no_grad():
            outputs = model(
                input_ids=feat["input_ids"].unsqueeze(0).to(DEVICE),
                bbox=feat["bbox"].unsqueeze(0).to(DEVICE),
                attention_mask=feat["attention_mask"].unsqueeze(0).to(DEVICE)
            )

        prob = torch.softmax(outputs.logits, dim=-1)
        pred_idx = torch.argmax(prob, dim=-1).item()

        # On utilise les clés originales : 'id', 'page', 'path'
        results.append({
            "doc_id": item['id'],
            "page_num": item['page'],
            "file_path": item['path'],
            "prediction": "DRUG_EVAL" if pred_idx == 1 else "OTHER",
            "confidence": f"{prob[0][pred_idx].item():.4f}"
        })

    # Écriture CSV
    with open(output_file, "w", newline="", encoding="utf-8") as f:
        writer = csv.DictWriter(f, fieldnames=["doc_id", "page_num", "file_path", "prediction", "confidence"])
        writer.writeheader()
        writer.writerows(results)

    print(f"✅ Rapport enregistré sous : {output_file}")

# --- LANCEMENT ---
doc_list = documents
generate_report(doc_list, model, tokenizer)
```

```python
import torch
import pytesseract
import csv
from PIL import Image
from pathlib import Path
from tqdm import tqdm
from pdf2image import convert_from_path
from transformers import LayoutLMForSequenceClassification, LayoutLMTokenizer

# --- CONFIGURATION ---
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MODEL_PATH = "./layoutlm_manual_final"
tokenizer = LayoutLMTokenizer.from_pretrained("microsoft/layoutlm-base-uncased")
model = LayoutLMForSequenceClassification.from_pretrained(MODEL_PATH).to(DEVICE)
model.eval()

# --- 1. FONCTION D'ALIGNEMENT MANUEL (Zéro Erreur de Dimension) ---
def prepare_manual_tokens(tokenizer, words, bboxes, max_len=512):
    token_ids, token_boxes = [], []
    for word, box in zip(words, bboxes):
        word_tokens = tokenizer.encode(word, add_special_tokens=False)
        token_ids.extend(word_tokens)
        token_boxes.extend([box] * len(word_tokens))

    # Troncature et ajout des tokens spéciaux [CLS] / [SEP]
    token_ids = [tokenizer.cls_token_id] + token_ids[:max_len-2] + [tokenizer.sep_token_id]
    token_boxes = [[0, 0, 0, 0]] + token_boxes[:max_len-2] + [[0, 0, 0, 0]]

    # Padding manuel strict à 512
    pad_len = max_len - len(token_ids)
    attention_mask = [1] * len(token_ids) + [0] * pad_len
    token_ids += [tokenizer.pad_token_id] * pad_len
    token_boxes += [[0, 0, 0, 0]] * pad_len

    return {
        "input_ids": torch.tensor(token_ids).unsqueeze(0).to(DEVICE),
        "bbox": torch.tensor(token_boxes).unsqueeze(0).to(DEVICE),
        "attention_mask": torch.tensor(attention_mask).unsqueeze(0).to(DEVICE)
    }

# --- 2. PRÉPARATION DES DOCUMENTS (VOTRE BOUCLE) ---
documents = []
i_c = 1
layout_dir = Path("data/layouts")
input_files = list(layout_dir.rglob("*.pdf"))

print(f"📦 Conversion de {len(input_files)} PDFs en images...")
for pdf_path in input_files:
    images = convert_from_path(pdf_path)
    for i, image in enumerate(images):
        img_name = f"{pdf_path.stem}_page_{i}.jpg"
        img_path = layout_dir / img_name
        image.save(img_path, "JPEG")
        documents.append({
            "id": i_c,
            "page": i,
            "path": str(img_path),
            "doc_path": str(pdf_path)
        })
    i_c += 1

# --- 3. INFERENCE ET REPORTING ---
final_results = []
print(f"🔍 Analyse LayoutLM sur {len(documents)} pages...")

for item in tqdm(documents):
    try:
        image = Image.open(item['path']).convert("RGB")
        w, h = image.size
        ocr = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)

        words, bboxes = [], []
        for j, text in enumerate(ocr['text']):
            if text.strip() and int(ocr['conf'][j]) > 15:
                box = [
                    max(0, min(int(1000 * (ocr['left'][j] / w)), 999)),
                    max(0, min(int(1000 * (ocr['top'][j] / h)), 999)),
                    max(0, min(int(1000 * ((ocr['left'][j] + ocr['width'][j]) / w)), 999)),
                    max(0, min(int(1000 * ((ocr['top'][j] + ocr['height'][j]) / h)), 999))
                ]
                words.append(text)
                bboxes.append(box)

        if not words: continue

        feat = prepare_manual_tokens(tokenizer, words, bboxes)

        with torch.no_grad():
            outputs = model(**feat)
            probs = torch.softmax(outputs.logits, dim=-1)
            pred_idx = torch.argmax(probs, dim=-1).item()

        final_results.append({
            "doc_id": item['id'],
            "page": item['page'],
            "file_name": Path(item['doc_path']).name,
            "prediction": "DRUG_EVAL" if pred_idx == 1 else "OTHER",
            "confidence": round(probs[0][pred_idx].item(), 4)
        })
    except Exception as e:
        print(f"⚠️ Erreur sur {item['path']}: {e}")

# --- 4. EXPORTATION CSV ---
output_file = "rapport_final_layoutlm.csv"
with open(output_file, "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["doc_id", "page", "file_name", "prediction", "confidence"])
    writer.writeheader()
    writer.writerows(final_results)

print(f"✅ Terminé ! Rapport disponible : {output_file}")
```

```python
import torch
import pytesseract
import nltk
from PIL import Image, ImageDraw, ImageFont
from transformers import LayoutLMForTokenClassification, LayoutLMTokenizer
from nltk.tokenize import word_punct_tokenize

# --- CONFIGURATION ---
MODEL_PATH = "./layoutlm_ner_final"
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Palette de couleurs pour les entités
COLOR_MAP = {
    "B-TOTAL": "red", "I-TOTAL": "orange",
    "B-DATE": "blue", "I-DATE": "cyan",
    "B-ID": "green", "I-ID": "lime",
    "O": None # On ne dessine pas les 'O' (Outside)
}

# 1. Chargement
tokenizer = LayoutLMTokenizer.from_pretrained("microsoft/layoutlm-base-uncased")
model = LayoutLMForTokenClassification.from_pretrained(MODEL_PATH).to(DEVICE)
model.eval()

# ID to Label (à adapter selon votre entraînement)
id2label = {0: "O", 1: "B-TOTAL", 2: "I-TOTAL", 3: "B-DATE", 4: "I-DATE", 5: "B-ID"}

def run_ner_inference(image_path):
    image = Image.open(image_path).convert("RGB")
    draw = ImageDraw.Draw(image)
    w, h = image.size

    # 2. OCR + NLTK (Segmentation fine)
    ocr = pytesseract.image_to_data(image, output_type=pytesseract.Output.DICT)
    words, bboxes = [], []

    for i, raw_word in enumerate(ocr['text']):
        if raw_word.strip() and int(ocr['conf'][i]) > 20:
            sub_tokens = word_punct_tokenize(raw_word)
            l, t, wd, ht = ocr['left'][i], ocr['top'][i], ocr['width'][i], ocr['height'][i]
            norm_box = [max(0, min(int(1000 * (l/w)), 999)), max(0, min(int(1000 * (t/h)), 999)),
                        max(0, min(int(1000 * ((l+wd)/w)), 999)), max(0, min(int(1000 * ((t+ht)/h)), 999))]

            for st in sub_tokens:
                words.append(st)
                bboxes.append(norm_box)

    # 3. Préparation Tenseurs (Alignement Sub-tokens)
    token_ids, token_boxes, token_actual_boxes = [], [], []
    # On garde les coordonnées réelles (l,t,w,h) pour le dessin final

    for word, box in zip(words, bboxes):
        word_tokens = tokenizer.encode(word, add_special_tokens=False)
        for t_id in word_tokens:
            token_ids.append(t_id)
            token_boxes.append(box)

    # Padding / Troncature à 512
    token_ids = [tokenizer.cls_token_id] + token_ids[:510] + [tokenizer.sep_token_id]
    token_boxes = [[0,0,0,0]] + token_boxes[:510] + [[0,0,0,0]]

    # 4. Prédiction
    input_ids = torch.tensor(token_ids).unsqueeze(0).to(DEVICE)
    bbox = torch.tensor(token_boxes).unsqueeze(0).to(DEVICE)

    with torch.no_grad():
        outputs = model(input_ids=input_ids, bbox=bbox)
        predictions = torch.argmax(outputs.logits, dim=-1).squeeze().tolist()

    # 5. Visualisation
    # On itère sur les tokens prédits (en sautant CLS et SEP)
    for i, pred_id in enumerate(predictions[1:-1]):
        label = id2label[pred_id]
        if label != "O":
            box = token_boxes[i+1]
            # Conversion 0-1000 vers pixels réels
            real_box = [box[0]*w/1000, box[1]*h/1000, box[2]*w/1000, box[3]*h/1000]

            color = COLOR_MAP.get(label, "yellow")
            draw.rectangle(real_box, outline=color, width=3)
            draw.text((real_box[0], real_box[1]-15), label, fill=color)

    image.show()
    image.save("ner_result.jpg")

# Test final
run_ner_inference(documents[0]['path'])
```
