```python
# Define the relative path
layout_dir = Path("data/layouts")
# 1. Grab all PDFs in your input folder
input_files = layout_dir.rglob("*.pdf")

for pdf_path in input_files:
    print(f"Processing: {pdf_path}")
    
    # 2. Convert PDF to image (Using the Poppler path from your setup)
    images = convert_from_path(pdf_path.resolve())
    # 3. Save each page as a JPEG
    for i, image in enumerate(images):
        image.save(f"{pdf_path}_page_{i}.jpg", "JPEG")
```

```python
INPUT_DIR = "data/layouts"
OUTPUT_FILE = "data/layouts/import_label_studio.json"
OUTPUT_DATASET_FILE = "data/layouts/import_label_studio.json"

def convert_to_ls(image_path):
    img = Image.open(image_path)
    w, h = img.size
    
    # Récupérer les données détaillées de Tesseract (dict)
    data = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)
    
    results = []
    
    for i in range(len(data['text'])):
        # On ne garde que les blocs avec du texte et une confiance > 0
        if int(data['conf'][i]) > 0 and data['text'][i].strip():
            # Conversion des pixels en pourcentages Label Studio
            x = (data['left'][i] / w) * 100
            y = (data['top'][i] / h) * 100
            width = (data['width'][i] / w) * 100
            height = (data['height'][i] / h) * 100
            
            # ID unique pour lier la BBox et le Texte
            region_id = f"region_{i}"
            
            # 1. Création de la Bounding Box
            results.append({
                "original_width": w,
                "original_height": h,
                "image_rotation": 0,
                "value": {
                    "x": x, "y": y, 
                    "width": width, "height": height, 
                    "rotation": 0, 
                    "rectanglelabels": ["Texte"]
                },
                "id": region_id,
                "from_name": "bbox",
                "to_name": "image",
                "type": "rectanglelabels"
            })
            
            # 2. Création de la Transcription (Text Area)
            results.append({
                "original_width": w, 
                "original_height": h,
                "value": {
                    "x": x, "y": y, 
                    "width": width, "height": height, 
                    "rotation": 0, 
                    "text": [data['text'][i]]
                },
                "id": region_id,
                "from_name": "transcription",
                "to_name": "image",
                "type": "textarea"
            })

    return {
        "data": {
            "ocr": f"/data/local-files/?d={image_path.replace(os.sep, '/')}"
        },
        "predictions": [{"model_version": "tesseract", "result": results}]
    }

# Main loop
all_tasks = []
for img_file in glob.glob(f"{INPUT_DIR}/*.jpg") + glob.glob(f"{INPUT_DIR}/*.png"):
    print(f"Traitement de : {img_file}")
    all_tasks.append(convert_to_ls(img_file))

with open(OUTPUT_FILE, 'w', encoding='utf-8') as f:
    json.dump(all_tasks, f, indent=2, ensure_ascii=False)

print(f"\nFichier prêt ! Importez '{OUTPUT_FILE}' dans Label Studio.")
```

```python
def normalize_bbox(bbox, width, height):
    """Normalise les coordonnées (x, y, w, h) sur une échelle de 0-1000."""
    return [
        int(1000 * (bbox[0] / width)),        # x_min
        int(1000 * (bbox[1] / height)),       # y_min
        int(1000 * ((bbox[0] + bbox[2]) / width)), # x_max
        int(1000 * ((bbox[1] + bbox[3]) / height))  # y_max
    ]
```

```python
def get_layoutlm_features(image_path):
    img = Image.open(image_path).convert("RGB")
    width, height = img.size
    
    # Extraction des données avec Tesseract
    ocr_data = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)
    
    words = []
    bboxes = []
    
    for i in range(len(ocr_data['text'])):
        conf = int(ocr_data['conf'][i])
        text = ocr_data['text'][i].strip()
        
        if conf > 0 and text:
            # Coordonnées brutes
            l, t, w, h = ocr_data['left'][i], ocr_data['top'][i], ocr_data['width'][i], ocr_data['height'][i]
            
            # Normalisation pour LayoutLM (0 à 1000)
            # Format attendu : [x_min, y_min, x_max, y_max]
            x0 = int(1000 * (l / width))
            y0 = int(1000 * (t / height))
            x1 = int(1000 * ((l + w) / width))
            y1 = int(1000 * ((t + h) / height))
            
            # Sécurité pour ne pas dépasser 1000
            bboxes.append([
                min(max(0, x0), 1000),
                min(max(0, y0), 1000),
                min(max(0, x1), 1000),
                min(max(0, y1), 1000)
            ])
            words.append(text)
            
    return {"words": words, "bboxes": bboxes, "image_path": image_path}

# Traitement par lots
dataset = []
for img_path in glob.glob(f"{INPUT_DIR}/*.jpg"):
    print(f"Normalisation de {img_path} pour LayoutLM...")
    dataset.append(get_layoutlm_features(img_path))

# Sauvegarde pour l'entraînement
with open("layoutlm_input.json", "w", encoding="utf-8") as f:
    json.dump(dataset, f, indent=2, ensure_ascii=False)
```

```python
import torch
from torch.utils.data import DataLoader
from transformers import LayoutLMForSequenceClassification, LayoutLMTokenizer, AdamW
import json

# 1. Configuration et Chargement
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_name = "microsoft/layoutlm-base-uncased"
tokenizer = LayoutLMTokenizer.from_pretrained(model_name)

# On suppose 2 classes : 0 = Facture, 1 = Autre
model = LayoutLMForSequenceClassification.from_pretrained(model_name, num_labels=2)
model.to(device)

with open("layoutlm_input.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# 2. Fonction de préparation des données
def encode_data(item, label):
    encoding = tokenizer(
        item["words"],
        boxes=item["bboxes"],
        return_tensors="pt",
        truncation=True,
        padding="max_length",
        max_length=512
    )
    return {
        "input_ids": encoding["input_ids"].flatten().to(device),
        "attention_mask": encoding["attention_mask"].flatten().to(device),
        "bbox": encoding["bbox"].flatten(0, 1).to(device),
        "labels": torch.tensor(label).to(device)
    }

# 3. Boucle d'entraînement simplifiée (Exemple)
optimizer = AdamW(model.parameters(), lr=5e-5)
model.train()

print(f"Entraînement lancé sur : {device}")

# Simulation d'une époque (à adapter avec vos vrais labels)
for item in data:
    optimizer.zero_grad()
    
    # Ici, '0' est un label fictif pour l'exemple
    inputs = encode_data(item, label=0)
    
    outputs = model(
        input_ids=inputs["input_ids"].unsqueeze(0),
        bbox=inputs["bbox"].unsqueeze(0),
        attention_mask=inputs["attention_mask"].unsqueeze(0),
        labels=inputs["labels"].unsqueeze(0)
    )
    
    loss = outputs.loss
    loss.backward()
    optimizer.step()
    print(f"Loss: {loss.item():.4f}")

# Sauvegarde du modèle
model.save_pretrained("./layoutlm_finetuned")
print("Modèle sauvegardé dans ./layoutlm_finetuned")
```
