```python
import os
import platform
from pathlib import Path
import glob
from pdf2image import convert_from_path
import pytesseract
from PIL import Image
```

```python
print(platform.machine(), platform.system(),platform.node(), platform.python_version(), platform.python_compiler(), platform.architecture())
```

```python

# Define the relative path
layout_dir = Path("data/layouts")
# 1. Grab all PDFs in your input folder
input_files = layout_dir.rglob("*.pdf")

for pdf_path in input_files:
    print(f"Processing: {pdf_path}")
    
    # 2. Convert PDF to image (Using the Poppler path from your setup)
    images = convert_from_path(pdf_path.resolve())
    
    # 3. Save each page as a JPEG
    for i, image in enumerate(images):
        image.save(f"{pdf_path}_page_{i}.jpg", "JPEG")
```

```python


# Charger l'image du formulaire (page du PDF convertie)
img = Image.open("page_formulaire.jpg")

# Récupérer les données détaillées (niveau mot)
data = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)

# 'data' contient maintenant :
# - 'text' : le mot reconnu
# - 'left', 'top' : coordonnées X, Y du coin supérieur gauche
# - 'width', 'height' : dimensions de la boîte
# - 'conf' : indice de confiance (0 à 100)
```
